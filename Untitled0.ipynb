```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
```

Data Preprocessing, cleaning,feature scaling

```python
df=pd.read_csv('advertising_sales_data.csv')
x=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(x[:,:])
x[:,:]=imputer.transform(x[:,:])
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
sc=StandardScaler()
sb=Normalizer()
b_train=sc.fit_transform(x_train[:,:])
b_test=sc.transform(x_test[:,:])
a_train=sb.fit_transform(x_train[:,:])
a_test=sb.transform(x_test[:,:])
```

This part answers the questions like
1.	What is the average amount spent on TV advertising in the dataset?
2.	What is the correlation between radio advertising expenditure and product sales?
3.	Which advertising medium has the highest impact on sales based on the dataset?


```python
print(df['TV'].describe())
numeric_df=df.select_dtypes(include=['int64','float64'])
cor_matrix=numeric_df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(cor_matrix,annot=True,cmap='PuBuGn',fmt='.2f')
plt.show()
```

linear regression model on the data

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(b_train,y_train)
print(regressor.coef_)
print(regressor.intercept_)
```

```python
y_pred=regressor.predict(b_test)
print(y_pred)
a=np.array([200,40,50])
c=regressor.predict(a.reshape(1,-1))
print(c[0])
print(df.head())
```

```python
reg2=LinearRegression()
reg2.fit(a_train,y_train)
y2_pred=reg2.predict(a_test)
print(y2_pred)
print(reg2.coef_)
print(reg2.intercept_)
```

```python
reg3=LinearRegression()
reg3.fit(x_train,y_train)
y3_pred=reg3.predict(x_test)
print(y3_pred)
print(y_test)
q=reg3.coef_
p=reg3.intercept_
print(200*q[0]+40*q[1]+50*q[2]+p)
```
